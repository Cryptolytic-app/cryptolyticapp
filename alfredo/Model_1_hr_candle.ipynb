```
def fill_nan(df):
  
  '''iterates through a dataframe and fills nans with appropriate open, high, low, close values'''
  
  df = df.copy()
  
  # forward filling the close column
  df['close'] = df['close'].ffill()
  
  # backward filling the open, high, low rows with the close value
  df = df.bfill(axis=1)
  
  return df
```

```
df3['close_diff'] = (df3['close'] - df3['close'].shift(1))/df3['close'].shift(1)
```

```
# function to create a target
def price_increase(x):
  if x > 0.001:
    return True
  else:
    return False
```

```
df4 = df3.fillna(method='ffill')
```
