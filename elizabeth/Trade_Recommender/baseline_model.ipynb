```
# TA library from github https://github.com/bukosabino/ta
df = add_all_ta_features(df, "open", "high", "low", "close", "volume")
```

```
# drop null values
null_columns = ['volume_obv', 'trend_adx', 'trend_adx_pos', 'trend_adx_neg', 'trend_trix']

df.drop(columns=null_columns, inplace=True)

df2 = df[40:]

df2['close_diff'] = df2['close'] - df2['close'].shift(1)
```

```
df2.shape
# shape should be (440043, 60)
```

```
df2.head()
```

```
# function to create a target
def price_increase(x):
  if x > 0:
    return True
  else:
    return False
```

```
# fix the index numbers
df2 = df2.reset_index()
df2.drop(columns='index', inplace=True)
```

```
# create target column
target = df2['close_diff'].apply(price_increase)

# remove the first item off target and remove the last item off the dataframe
target3 = target[1:].values
df3 = df2[:440042]

# create target column
df3['target'] = target3
```

```
df3.head(10)
```

```
# export to csv
df3[1:].to_csv('BTC_USD_CBPro_TA.csv', index=False)
```

## Baseline Model

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```
# load data
traintest = pd.read_csv('BTC_USD_CBPro_TA.csv')

train = traintest[traintest['time'] < 1538265600] # cutoff sept 30 2018
test = traintest[traintest['time'] > 1548892800] # cutoff jan 31 2019
train.shape, test.shape
```

```
# pick features and target
features = ['open', 'high', 'low', 'close', 'volume']
target = 'target'

# X, y vectors
X_train = train[features]
X_test = test[features]
y_train = train[target]
y_test = test[target]

# select model
model = RandomForestClassifier()

# fit model
model.fit(X_train, y_train)
```

```
# make predictions 
y_pred = model.predict(X_test)

# accuracy
score = accuracy_score(y_test, y_pred)
print(score)
```

## Model Tuning

```
# pick features and target
feature_columns = traintest.drop(columns=['target', 'time']).columns.tolist()
features = feature_columns
target = 'target'

# X, y vectors
X_train = train[features]
X_test = test[features]
y_train = train[target]
y_test = test[target]

# select model
model = RandomForestClassifier(max_depth=20, n_estimators=100, n_jobs=-1)

# fit model
model.fit(X_train, y_train)
```

```
# make predictions 
y_preds = model.predict(X_test)

score = accuracy_score(y_test, y_preds)
print(score)
```
