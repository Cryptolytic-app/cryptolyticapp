```python
import websocket
import json
from pprint import pprint
import time
import pandas as pd

URL = "wss://ws-feed.pro.coinbase.com"

df = pd.DataFrame()

def on_message(ws, message):
    """Callback executed when a message comes.
    Positional argument:
    message -- The message itself (string)
    """
    msg = json.loads(message)
    pprint(msg)
    global df
    if msg['type'] == 'match':
        df = df.append(msg, ignore_index=True)
        
def on_open(socket):
    """Callback executed at socket opening.
    Keyword argument:
    socket -- The websocket itself
    """
    params = {
        "type": "subscribe",
        "channels": [{"name": "matches", "product_ids": ["BTC-USD"]}]
    }
    socket.send(json.dumps(params))

def main():
    """Main function."""
    ws = websocket.WebSocketApp(URL, on_open=on_open, on_message=on_message)
    ws.run_forever()
      
if __name__ == '__main__':
    main()
```

```python
new_index = pd.DatetimeIndex(df['time'])
df = df.set_index(new_index).drop(columns='time')
df['price'] = df['price'].astype(float)
ohlc = df['price'].resample('1Min').ohlc()
ohlc
```
