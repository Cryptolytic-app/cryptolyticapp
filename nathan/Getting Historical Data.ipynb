```python
import requests
import pandas as pd
from tqdm import tqdm
import time
from IPython.display import display

start_of_oct_2019 = 1569888000
start_of_2015 = 1420070400

base_url = 'https://api.bitfinex.com/v2/candles/trade:'

# see https://api-pub.bitfinex.com/v2/tickers?symbols=ALL
currency_pair = 'STJBTC'
interval = '5m'
limit = 1000
end = start_of_oct_2019
start = end - (5*60*999)

r = requests.get(base_url +
                 interval +
                 ':t' +
                 currency_pair +
                 '/hist?start=' +
                 str(start*1000) +
                 '&end=' + 
                 str(end*1000) +
                 '&limit=' +
                 str(limit))

columns = ['time', 'open', 'close', 'high', 'low', 'volume']
df = pd.DataFrame(r.json(), columns=columns)

pbar = tqdm(total=((start-start_of_2015)//(5*60*1000)))
while end >= start_of_2015:
    end = start - (5*60)
    start = end - (5*60*999)
    r = requests.get(base_url +
                     interval +
                     ':t' +
                     currency_pair +
                     '/hist?start=' +
                     str(start*1000) +
                     '&end=' + 
                     str(end*1000) +
                     '&limit=' +
                     str(limit))
    to_append = pd.DataFrame(r.json(), columns=columns)
    df = df.append(to_append)
    df = df.reset_index(drop=True)
    time.sleep(1.25)
    pbar.update(1)
pbar.close()

display(df)

df.to_csv('bitfinex_' + currency_pair.lower() + '.csv')
```

```python
import requests
import datetime as DT
import pandas as pd
from tqdm import tqdm
import time
from IPython.display import display

start_of_oct_2019 = 1569888000
start_of_2015 = 1420070400

base_url = 'https://api.pro.coinbase.com/products/'

# see https://api.pro.coinbase.com/products/
currency_pair = 'XTZ-USD'
interval = 300
end = start_of_oct_2019
start = end - (5*60*299)

r = requests.get(base_url +
                 currency_pair +
                 '/candles/?granularity=' +
                 str(interval) +
                 '&start=' +
                 DT.datetime.utcfromtimestamp(start).isoformat() +
                 '&end=' + 
                 DT.datetime.utcfromtimestamp(end).isoformat())

columns = ['time', 'low', 'high', 'open', 'close', 'volume']
df = pd.DataFrame(r.json(), columns=columns)

pbar = tqdm(total=((start-start_of_2015)//(5*60*300)))
while end >= start_of_2015:
    end = start - (5*60)
    start = end - (5*60*299)
    r = requests.get(base_url +
                     currency_pair +
                     '/candles/?granularity=' +
                     str(interval) +
                     '&start=' +
                     DT.datetime.utcfromtimestamp(start).isoformat() +
                     '&end=' + 
                     DT.datetime.utcfromtimestamp(end).isoformat())
    to_append = pd.DataFrame(r.json(), columns=columns)
    df = df.append(to_append)
    df = df.reset_index(drop=True)
    time.sleep(.75)
    pbar.update(1)
pbar.close()

display(df)

df.to_csv('coinbase_' + currency_pair.lower().replace('-', '_') + '.csv')
```

```python
import requests
import pandas as pd
from tqdm import tqdm
from IPython.display import display

start_of_oct_2019 = 1569888000
start_of_2015 = 1420070400

base_url = ("https://poloniex.com/public?command=returnChartData"
            "&currencyPair=")

# see https://poloniex.com/public?command=returnTicker
currency_pair = 'BTC_LTC'
interval = (5*60)
end = start_of_oct_2019
start = end - (60*60*24*10)

r = requests.get(base_url +
                 currency_pair +
                 '&start=' +
                 str(start) +
                 '&end=' + 
                 str(end) +
                 '&period=' +
                 str(interval))

columns = ['date', 'open', 'close', 'high', 'low', 'volume']
df = pd.DataFrame(r.json())[columns]

pbar = tqdm(total=((start-start_of_2015)//(60*60*24*10)))
while end >= start_of_2015:
    end = start - (5*60)
    start = end - (60*60*24*10)
    r = requests.get(base_url +
                     currency_pair +
                     '&start=' +
                     str(start) +
                     '&end=' + 
                     str(end) +
                     '&period=' +
                     str(interval))
    to_append = pd.DataFrame(r.json())[columns]
    df = df.append(to_append)
    df = df.reset_index(drop=True)
    pbar.update(1)
pbar.close()

df = df[df['date'] != 0]
display(df)
    
df.to_csv('poloniex_' + currency_pair.lower() + '.csv')
```
