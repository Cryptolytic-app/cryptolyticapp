```python
import requests

response = requests.get('https://api.pro.coinbase.com/products/')
response.json()

coinbase_pairs = [entry['id'] for entry in response.json()]
coinbase_pairs.sort()

coinbase_pairs = [pair.replace('-', '_') for pair in coinbase_pairs]
coinbase_pairs
```

```python
response = requests.get('https://api-pub.bitfinex.com/v2/tickers?symbols=ALL')
response.json()

bitfinex_pairs = [entry[0][1:] for entry in response.json() if 
                  entry[0].startswith('t')]

sixes = [pair for pair in bitfinex_pairs if len(pair) == 6]
sixes = [(pair[:3] + '_' + pair[3:]) for pair in sixes]

other = [pair for pair in bitfinex_pairs if len(pair) != 6]
other = [pair.replace(':', '_') for pair in other]

bitfinex_pairs = sixes + other
bitfinex_pairs.sort()
bitfinex_pairs
```

```python
response = requests.get('https://poloniex.com/public?command=returnTicker')
response.json()

poloniex_pairs = list(response.json().keys())
poloniex_pairs.sort()

poloniex_pairs
```
